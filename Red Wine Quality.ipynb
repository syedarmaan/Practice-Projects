```python
bins = (2, 6.5, 8)
groups = ['bad', 'good']
df['quality'] = pd.cut(df['quality'], bins = bins, labels = groups)
```

```python
label_quality = LabelEncoder()
```

```python
df['quality'] = label_quality.fit_transform(df['quality'])
```
